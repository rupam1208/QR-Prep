```python
import numpy as np
import importlib
import utils.finance_tools as ft
import matplotlib.pyplot as plt
import random
```

```python
def asset_price_gm(p,pr,it=3000):
    for i in range(it):
        if random.random()<pr:
            p=p*1.01;
        else:
            p=p*0.99;
    return p;
```

```python
def asset_price_am(p,pr,it=3000):
    for i in range(it):
        if random.random()<pr:
            p=p+1;
        else:
            p=p-1;
    return p;
```

```python
asset_price_gm(100,0.55)
asset_price_gm(100,0.45)
```

```python
asset_price_am(100,0.55)
asset_price_am(100,0.45)
```

```python
sp=100;freq=10000+1;
plt.plot(range(1,freq),np.array([asset_price_gm(sp,0.49,it=i) for i in range(1,freq)]))
plt.plot(range(1,freq),np.array([asset_price_gm(sp,0.50,it=i) for i in range(1,freq)]))
plt.plot(range(1,freq),np.array([asset_price_gm(sp,0.51,it=i) for i in range(1,freq)]))
plt.plot(range(1,freq),np.array([asset_price_gm(sp,0.52,it=i) for i in range(1,freq)]))
plt.yscale('log')
plt.xlabel("Time")
plt.ylabel("Stock price")
```

```python
sp=100;freq=10000+1;
plt.plot(range(1,freq),np.array([asset_price_am(sp,0.49,it=i) for i in range(1,freq)]))
plt.plot(range(1,freq),np.array([asset_price_am(sp,0.50,it=i) for i in range(1,freq)]))
plt.plot(range(1,freq),np.array([asset_price_am(sp,0.51,it=i) for i in range(1,freq)]))
plt.plot(range(1,freq),np.array([asset_price_am(sp,0.52,it=i) for i in range(1,freq)]))
plt.xlabel("Time")
plt.ylabel("Stock price")
```
