```python
s=np.linspace(1,200)
```

```python
def plot_payoff(func,S,K,P,*args,label=None): 
    """Plots payoff functions according to func automatically"""
    plt.plot(S,func(S,K,P),*args,label=label)
    plt.xlabel("S")
    plt.ylabel("Payoff")
    plt.axhline(0, color="black", linestyle='--')
    plt.xlim(0,np.max(S)/2)
    plt.ylim(-np.max(S)/2,np.max(S)/2)
    plt.legend()
```

```python
#Forward contract (Same for futures contract assuming it is not closed out)

plt.plot(s,ft.fwd_pay_long(s,20), label=r"Long")
plt.plot(s,ft.fwd_pay_short(s,20), label=r"Short")
plt.xlabel("S")
plt.ylabel("Payoff (forward contract)")
plt.axhline(0, color="k")
plt.xlim(0,100)
plt.ylim(-100,100)
plt.legend()
plt.grid()
```

```python
#Options 

plt.plot(s,ft.opn_pay_long_call(s,40,15), label="Long call")
plt.plot(s,ft.opn_pay_short_call(s,40,15), label="Short call")
plt.plot(s,ft.opn_pay_long_put(s,40,15), label="Long put")
plt.plot(s,ft.opn_pay_short_put(s,40,15), label="Short put")
plt.xlabel("S")
plt.ylabel("Payoff (Option)")
plt.axhline(0,color="k")
plt.xlim(0,100)
plt.ylim(-100,100)
plt.legend(loc='best')
```
